## 1.3 Function<a id='1.3_function'></a>

```python
# transform a time series dataset into a supervised learning dataset
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
	#n_vars = 1 if type(data) is list else data.shape[1]
	df1 = pd.DataFrame(data)
	cols = list()
	# input sequence (t-n, ... t-1)
	for i in range(n_in, 0, -1):
		cols.append(df1.shift(i))
	# forecast (output) sequence (t, t+1, ... t+n)
	for i in range(0, n_out):
		cols.append(df1.shift(-i))
	# put it all together
	agg = pd.concat(cols, axis=1)
	# drop rows with NaN values
	if dropnan:
		agg.dropna(inplace=True)
	return agg.values
```

#### 1.3.1 Function: walk_forward_validation <a id = '1.3.1_walk_forward_validation'></a>

```python
def walk_forward_validation(data, n_test, n_out=1):
    predictions = list()
	# split dataset
    train, test = train_test_split(data, n_test) # np arrays
	# seed history with training dataset
    history = [x for x in train] # array to list (why list? Easier to append?)
	# step over each time-step in the test set
    for i in range(len(test)):
		# split test row into input and output columns
        testX, testy = test[i, :-n_out], test[i, -n_out:]
		# fit model on history and make a prediction
        yhat, modelXG = xgboost_forecast(history, testX, n_out)
		# store forecast in list of predictions
        predictions.append(yhat)
		# add actual observation to history for the next loop
        history.append(test[i])
		# summarize progress
        print('>expected=%.1f, predicted=%.1f' % (np.mean(testy), np.mean(yhat)))	
    # estimate prediction error
    error = mean_absolute_error(test[:, -n_out:], predictions)
    return error, test[:, -n_out:], predictions, modelXG
```

```python
# fit an xgboost model and make a one step prediction
def xgboost_forecast(train, testX, n_out=1):
	# transform list into array
	train = np.asarray(train)
	# split into input and output columns
	trainX, trainy = train[:, :-n_out], train[:, -n_out:]
	# fit model
	model = XGBRegressor(objective='reg:squarederror', n_estimators=1000)
	model.fit(trainX, trainy)
	# make a one-step prediction
	yhat = model.predict(np.asarray([testX]))
	return yhat[0], model
```

```python
def train_test_split(data, n_test):
    return data[:-n_test,:], data[-n_test:,:]
```

```python
def create_lagged_df(dataframeToLag, n_in=1, n_out=1, dropnan=False):
    dataframeToLag = pd.DataFrame(dataframeToLag)
    df_lagged = pd.DataFrame()
    df_lagged['date'] = dataframeToLag.index
    for col in dataframeToLag.columns:
        lag_arr = series_to_supervised(dataframeToLag[col], n_in, n_out, dropnan=False)
        lag_columns = [(col+'_'+str(lag)) for lag in range(n_in,-n_out,-1)]
        for lag in range(0,n_in+n_out):
            df_lagged[lag_columns[lag]]=lag_arr[:,lag]
    df_lagged.set_index('date', inplace=True)
    return df_lagged
```

```python
# load the saved dictionary from pickle file
filePath_pickle = Path('/Users/parkj/Documents/pyDat/dataSet/covid_countryData.pickle')
pickle_in = open(filePath_pickle, 'rb')
country_dict = pickle.load(pickle_in)
df = country_dict['United States']
```

```python
# time series (non-case) features 
features_lag = ['rtrc', 'grph', 'prks', 'tran', 'work', 'resi', 'vac', 'dayow']
# evaluate
n_test = 100
# non-case features (mobility, vaccination, dayow) with lagging for previous 7 days
df_feat_non_case_lag = create_lagged_df(df.loc[:,features_lag], n_in=7, n_out=0, dropnan=False)
# lagged case features for previous 7 days
df_feat_case_lag = create_lagged_df(df.case_mil_percMax, n_in=7, n_out=0, dropnan=False)
# join features 
df_feat_lag = df_feat_non_case_lag.join(df_feat_case_lag, on='date', how='left')
# target variables for next 7 days
df_targ_7day = create_lagged_df(df.case_mil_percMax, n_in=0, n_out=7, dropnan=False)
```

```python
# build models to predict each of the next 7 days
rez_XGBoost_dict = {}
for i in range(7):
    df_targ_this = df_targ_7day.drop(df_targ_7day.columns[[j for j in range(7) if j not in [i]]], axis=1, inplace=False)
    # join the feature and target data frames 
    df_feat_targ = df_feat_lag.join(df_targ_this, on='date', how='left')
    # drop NaNs
    df_feat_targ.dropna(inplace=True)
    # fit XGBoost model with walk-forward-validation
    error, y, yhat, modelXG = walk_forward_validation(df_feat_targ.values, n_test, n_out=1) # n_test = 100
    # dataframe y and yhat
    df_y_yhat = pd.DataFrame({'date':df_feat_targ.index[-n_test:], 'y': [float(i) for i in y], 'yhat': yhat})
    # bundle up the results
    rez_XGBoost_dict[df_targ_this.columns[0]] = {'error':error, 'y_yhat':df_y_yhat, 'model':modelXG}
```

## 1.3 <a id = "1.3_section"></a>

```python
# set up to draw model results
fig, ax = plt.subplots(7, 1, figsize=(12, 24))### 1.8.1 Plot Actual (Y) and Predicted (Yhat) Cases<a id='1.8.1_Plot_Actual_and_Predicted_Cases'></a>
fig.subplots_adjust(wspace=0.5, hspace=0.7)
# plot for each of the 7-day forecast
for i, key in enumerate(rez_XGBoost_dict.keys()):
    rez = rez_XGBoost_dict[key]['y_yhat']
    ax[i].plot(rez['date'], rez['y'], 'o-')
    ax[i].plot(rez['date'], rez['yhat'], 'o-')
    ax[i].set(xlabel='Date', ylabel='Percent Cases', title=key)
    ax[i].legend(['y','yhat'], prop=dict(weight='bold',size=14))
```
